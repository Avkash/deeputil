# API to load VGG16 and VGG19 ImageNet Models

You just need to call the following API:
    
 - ModelsMaster.get_vgg16_model
 - ModelsMaster.get_vgg19_model

```python
import tensorflow
import keras
from deeputil.model import ModelsMaster
from deeputil.modelassist import KerasModelHelper
from deeputil.modelassist import ImportExport
```

```python
vgg16_model = ModelsMaster.get_vgg16_model()
```

```python
KerasModelHelper.get_model_layers_details_all(vgg16_model)
```

```python
vgg19_model = ModelsMaster.get_vgg19_model(show_info=False)
```

```python
KerasModelHelper.get_model_layers_details_all(vgg19_model)
```

```python
# Now we are loading the VGG19 Model with no Top
vgg19_model_noTop = ModelsMaster.get_vgg19_model(show_info=False, include_top=False)
```

```python
# As you can see below, the flatten and dense layers are not loaded:
KerasModelHelper.get_model_layers_details_all(vgg19_model_noTop)
```

```python
#Now you can use these model as any keras models
assert isinstance(vgg19_model_noTop, keras.engine.training.Model)
assert isinstance(vgg19_model, keras.engine.training.Model)
assert isinstance(vgg16_model, keras.engine.training.Model)
```
